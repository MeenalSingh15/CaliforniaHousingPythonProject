6. Write a programming construct (create a user defined function) to calculate the median value of the data set wherever required.

```python
def calculate_median(dataset, column_name):
    """
    Calculate the median value of a specified column in a DataFrame.

    Parameters:
    - data: DataFrame, the input dataset
    - column_name: str, the name of the column for which to calculate the median
"""

    median_value = dataset[column_name].median()
    return median_value
    #Returns:- median_value:the calculated median value



# Specify the column for which you want to calculate the median for eg lets calculate median_income median value
column_of_interest = m['median_income']

# Call the user-defined function to calculate the median
median_result = calculate_median(m, 'median_income')

# Display the result
print(f"Median value of '{column_of_interest}': {median_result}")
```

10. Please create a new column named total_bedroom_size. If the total bedrooms is 10 or less, it should be quoted as small. If the total bedrooms is 11 or more but less than 1000, it should be medium, otherwise it should be considered large.



```python
# Define a function to categorize total bedrooms
def categorize_bedroom_size(total_bedrooms):
    if total_bedrooms <= 10:
        return 'small'
    elif 10 < total_bedrooms < 1000:
        return 'medium'
    else:
        return 'large'

# Create the new column 'total_bedroom_size' using the apply method
m['total_bedroom_size'] = m['total_rooms'].apply(categorize_bedroom_size)

# Display the DataFrame with the new column
print(df)
```
